```python
# melt : convert wide data to long data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "District" : ["Jhrsuguda", "Sundargarh", "Sambalpur", "Deogarh"],
    "area" : [15000, 2000, 3000, 1000],
    "year" : [1900, 1901, 1902, 1903],
    "market" : [1200, 455, 6700, 5000],
    # "tax" : [20, 30, 25, 15]
})
```

```python
melted_df = df.melt(
    id_vars = ["District", "area"], # Those column are as it is
    value_vars = ["year", "market"], # Rest columns
    var_name = "Others", # column name store under Others
    value_name = "Value" 
)

melted_df
```

```python
# pivot : Convert long data to wide
original_df = melted_df.pivot(
    index = ["District", "area"],
    columns = "Others",
    values = "Value"
)

original_df
```
