```python
# Feature Engineering

# Transform data to create some new column, new features that improve the performance of our model
```

```python
import pandas as pd
```

```python
df = pd.read_csv("pd_employee_data.csv")
```

```python
df
```

```python
# Create a new column where we store the tax amount

# # .apply() - Write the logic inside a function than pass that function in apply()
```

```python
df2 = df.copy()

df2["tax"] = df2["income"].apply(lambda x : "20%" if x >= 50000 else "10%")
# here apply function is higher order function bcz anather funcyion is inside this function

df2
```

```python
# map() - map some value with existing value

# Convert female to F, male to M and Unknown to U
```

```python
gender_map = {
    "Male" : "M",
    "Female" : "F",
    "Unknown" : "U"
}

df2["gender"] = df2["gender"].map(gender_map)
```

```python
df2
```

```python
# assign() - By this we create a new column by assigning some value
df2.assign(new_income = df2["income"] * 1.1) # 1.1 for increase by 10%
```

```python
# Replace(old,new)
# Replace the country USA to US
```

```python
df2["country"] = df2["country"].replace("USA", "US")
```

```python
df2
```

```python
# rename( columns = {old : new})
```

```python
# Rename all column (here change to first letter Uppar case)
df2.columns = ["ID", "Name", "Age", "Country", "gender", "Income", "Tax"]
```

```python
df2
```

```python
# Rename specific column
df2.rename( columns = {"Age":"Age Value"})
```

```python
# Rename label
df2.rename(index = {0:"First"})
```

```python
# rank() : We can assign rank from lowest to heighst value
# Give rank on the basis of income
sorted_df["Ranking"] = sorted_df["Income"].rank() # if we pass ascending = false then it rank alternately
# If we want a whole value in each ranking then we use:  method = "dense" : here 4 be print
# If we want a min value of same rank then we use:  method = "min" : here 4 be print
# If we want a max value of same rank then we use:  method = "max" : here 5 be print
# When two value are same it print the average value like 4.5 in output
sorted_df
```
