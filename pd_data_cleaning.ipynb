```python
# Handle missing values
# code below
```

```python
# isnull - It returns true where null value present
df.isnull()
```

```python
# isna - It returns true where null value present
# The other nae of isnull
# The work of isnull and isna is same
df.isna()
```

```python
# isnull().sum() - Count the missing value per column
df.isnull().sum()
```

```python
# dropna - use to drop all the null value containing row from our data
df.dropna()
```

```python
# dropna(asis=1) - use to drop all the null value containing column from our data
df.dropna(axis=1)
```

```python
# fillna(value) - fill the missing place with passing value
df.fillna("Pritam")
```

```python
# Fill the missing value of age column by average value of present value
df["age"].fillna(df["age"].mean())
```

```python
# Store the clean data (age) in separate
cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)
cleaned_data
```
