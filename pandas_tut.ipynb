```python
import pandas as pd
```

```python
# data frame
# 2D labelled array

info = {
    "name" : ["Pritam", "Sritam", "Pritesh", "Sakhi"],
    "CGPA" : [9.5, 8.5, 9.3, 9.2]
    
}
# Convert this dict into data frame by DataFrame()
df = pd.DataFrame(info)
print(df)

# It shows us like a table, keys = header of the table
# In out put the numbering shows (0,1,2,3), these numbers are the lables, automatically created.

# Print all columns
print(df.index)

# print the index
print(df.columns)

# data frame is also be a list of list
animal_prop = pd.DataFrame([["Tiger", 4], ["Cat", 4], ["duck", 2]], columns = ["Animal", "Legs"])
print(animal_prop)
```

```python
# We can use list of list and data frame in numpy array also
import numpy as np
np_arr = np.array([[1,2,3],[4,5,6]])
df = pd.DataFrame(np_arr, columns=["A", "B", "C"])
print(df)
```

```python
# Series
# 1D labelled array

nums = [1,2,3,4,5,6]
# Convert this list into series by Series()
s = pd.Series(nums)
print(s)

# We can access indivisual value also by index based and labeled based
# Index based access
print(s[0])
print(s[2])
```

```python
# We can also define our custome labels
age_val = [25, 24,16, 20]

s = pd.Series(age_val, index = ["Pritam", "Sritam", "Pritesh", "Sakhi"])
print(s)

# Print all labeled value
print(s.index)

# Access value by babel, label based access
print(s["Pritam"]) # 25
print(s["Sakhi"]) #20
```

```python
# Series proporties
# 1 : Series are homogenous (Same type of data)
nums = [1,2,"ABC"] # It is not recomended
s = pd.Series(nums)
print(s)
# data type is object

# 2 : Vectorizzed operations.
nums1 = [1,2,3,4,5,6]
nums2 = [4,5,3,6,7,4]

s1 = pd.Series(nums1)
s2 = pd.Series(nums2)

print(s1 + s2)

# 3 : handle missing value with NaN
nums1 = [1,2,3,4,5,6]
nums2 = [4,5,3,6]

s1 = pd.Series(nums1)
s2 = pd.Series(nums2)

print(s1 + s2)

# 4 : Mutable values, immutable size
nums = [1,2,3,4,5,6]
s1 = pd.Series(nums)
s1[0] = 100
s1[5] = 600
print(s1)

# drop the first element
nums = [111,222,333]
s1 = pd.Series(nums)

changed_s1= s1.drop(0)
print(s1)
print(changed_s1)

# All these proporties are also work in data frame also
```
