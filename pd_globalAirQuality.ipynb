```python
# Chaining
# When we doing operation like slicing then it create a view of that.
# If we want to change any valuue then it change in the original data.
# So for this issue we use chaining by .copy() method.

cities = df["city"].copy()
cities[0] = "Mumbai"
print(cities)
# It's don't change in the original file by using .copy()
# The below o/p is copy, not the original file
```

```python
# Filtering data
# Give the details of those aqi value is more that 100
df[df["aqi"] > 100]

# We get our data that is 9721 as shown as below
```

```python
# Give the details of those aqi value is more that 100 and tempereture is greater that 30 degree
df[(df["aqi"] > 100) & (df["temperature"] > 30)]

# We get our data that is 2384 as shown as below
```

```python
# If you don't need all column
# We need only city name of those aqi value is more that 100 and tempereture is greate then 30

df[(df["aqi"] > 100) & (df["temperature"] > 30)] [["city", "aqi", "temperature"]]

# Pandas do not change the origibnl label at the time of filtering data.

# we can print the first row of this filtering data by label and incex both
# If we print by index then we use index 0
# If we print by label then we have to use 6 (label number)
```

```python
# Query method to filtering data.
# It is the easy way to filtering data. 
# Filter the data tjose aqi greater than 100 and temp greter thn 30
aqi_val = 100
df.query("aqi > @aqi_val & temperature > 30") [["city", "aqi", "temperature"]]

# rules
# Exprssion as a string
# column based reference
# Backtiks for col_name with space or special characters
# operators ( &, |, ~, >=, >, <, <=, == etc  )
# @ to reference python vars
# Chained technique for get specific cols.

# Query create copy

# No need to remember where copy and where view created, we just use .copy() every where.
```
