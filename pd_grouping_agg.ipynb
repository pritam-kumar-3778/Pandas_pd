```python
# Grouping & Aggregation of Data
```

```python
# First group the data based on something then do aggregation (mean/max/min etc)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("pd_employee_data.csv")
```

```python
df
```

```python
# Group the data based on country
df.groupby("country")["income"].min()

# Which Country having minimum income
```

```python
# For calculate multiple opr like mean, max and min 
df.groupby("country") ["income"].agg(["mean", "max", "min"])

# we can use agg() or aggregate(), both arr same.
```

```python
# Change column name
df.groupby("country") ["income"].agg(Avg_Salary = "mean", Max_Salary = "max", Min_Salary = "min")
```

```python
# Different opr based on different col value
df.groupby("country").agg({
    "income" : "mean",
    "age" : "min"
})
```

```python
# Change column name
df.groupby("country").agg(
    Avg_income = ("income","mean"),
    min_age = ("age", "min")
)
```
